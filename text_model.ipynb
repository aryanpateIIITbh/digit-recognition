```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
test_data = pd.read_csv("digit_test.csv")
```

```python
test_data = np.reshape(test_data.values, (test_data.shape[0], 28,28))
test_data = test_data.reshape(test_data.shape[0], test_data.shape[1], test_data.shape[2],1)
print("New shape of train data: ", test_data.shape)
```

```python
from tensorflow.keras.models import load_model 
model = load_model("digit1.h5") 
test_prediction  = model.predict(test_data)
test_prediction = np.argmax(test_prediction,axis=1)
```

```python
width=5
height=5
rows = 5
cols = 5
axes=[]
fig=plt.figure(figsize=(10,10))

for i in range(0,rows*cols):
    seven = test_data[i]
    seven = seven.reshape(28, 28)
    axes.append( fig.add_subplot(rows, cols, i+1) )
    plt.imshow(seven)
    plt.text(3,5,str(test_prediction[i]),fontsize=15,bbox={'alpha':1,'pad':9})
fig.tight_layout()    
plt.show()
```
